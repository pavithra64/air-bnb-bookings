```python
#Updating traces and layout
fig.update_traces(textfont=dict(size=15,color='black'))
fig.update_layout(title='Host has the highest number of reviews')
fig.show()
```

#___Which is the most cheapest area? Show the prize distribution.___

```python
#Extracing the columns which is required for analysis of area
new_airbnb_df=airbnb_df_new.loc[:,['neighbourhood_group','neighbourhood','price',]]
```

```python
#Grouping the Neighborhood group of cheapest area and perofrm mean aggregation function
result_bronx=new_airbnb_df.groupby(['neighbourhood_group','neighbourhood'])['price'].mean().reset_index()
result_bronx['price']=result_bronx['price']
result_bronx=result_bronx.loc[result_bronx['neighbourhood_group']=='Bronx']
result_bronx=result_bronx[['neighbourhood','price']]
```

```python
#Sorting the results
result_bronx.sort_values('price',ascending=False,inplace=True)
```

```python
#Computing Cumulative Percentage
result_bronx['cum_percent'] = 100*(result_bronx['price'].cumsum() / result_bronx['price'].sum())
```

```python
#plotting bar graph price vs neighbourhood_group
trace1 = go.Bar(
    x=result_bronx['neighbourhood'],
    y=result_bronx['price'],text=result_bronx['price'],
    name='price')
```

```python
#plotting Scatter graph cumulative percent vs neighbourhood_group
trace2 = go.Scatter(
    x=result_bronx['neighbourhood'],y=result_bronx['cum_percent'], name='cum_percent',yaxis='y2')
```

```python
#Merging the two plots to get parreto chart and finding the most cheapest area
fig = make_subplots(specs=[[{"secondary_y": True}]])
```

```python
#Updating traces and layout to beautify the plot and setting the font size
fig.add_trace(trace1)
fig.add_trace(trace2,secondary_y=True)
fig['layout'].update(height = 800, width = 1600,xaxis=dict(
      tickangle=-90
    ))
fig.update_layout(title='Bronx is Cheapest area and Average Price distribution of that area',plot_bgcolor='black',xaxis=dict(tickfont = dict(size=14)))
#show figure
fig.show()
```

From above plot we infer that:

***___Hunts point___ is the ___cheapest___ one with price of room is 50.5.***

___Riverdale___ is the ___most expensive___ one in the ___Bronx___ with average price of room is 442.09. which contribute 9.84% of price distribution from total Bronx neighborhood place of 100% price distribution.


___Riverdale,City Island,Spuyten Duyvil,Eastchester,Unionport,Westchester square and West Farms___ cumulative price percentage is __28.79%__ which lies in total Bronx neighborhood place price distribution of __100%__.
